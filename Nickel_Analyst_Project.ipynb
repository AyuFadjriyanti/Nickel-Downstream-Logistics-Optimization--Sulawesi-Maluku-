DATA CLEANING

```python
# Clean the column
mines.columns = mines.columns.str.strip()
ports.columns = ports.columns.str.strip()

mines.rename(
    columns={
        "keteragan": "keterangan"
    },
    inplace=True
)
```

```python
# Clean typo
mines.rename(
    columns={
        "keteragan": "keterangan"
    },
    inplace=True
)
```

```python
# Change ID columns to str
id_cols = [
    "Nomor Lokasi Keterdapatan",
    "objectid"
]

for col in id_cols:
    mines[col] = (
        mines[col]
        .astype(str)
        .str.replace(".", "", regex=False)
        .str.strip()
    )

mines["Nomor Lokasi Keterdapatan"] = (
    mines["Nomor Lokasi Keterdapatan"]
    .astype("string")
)
```

```python
# Cleaning big numeric columns
numeric_cols = [
    "Bijih Terbukti (Ton)",
    "Logam Terbukti (Ton)"
]

for col in numeric_cols:
    mines[col] = (
        mines[col]
        .astype(str)
        .str.replace(".", "", regex=False)
        .str.replace(",", ".", regex=False)
    )

    mines[col] = pd.to_numeric(
        mines[col],
        errors="coerce"
    )
```

```python
# Cleaning coordinate columns
coord_cols = [
    "longitude",
    "latitude"
]

for col in coord_cols:
    mines[col] = pd.to_numeric(
        mines[col],
        errors="coerce"
    )
```

```python
# Cleaning distance columns
distance_cols = [
    "Mine_port(meter)",
    "Mine_road(meter)",
    "Mine_Smelter(meter)",
    "Luas Wilayah"
]

for col in distance_cols:
    mines[col] = (
        mines[col]
        .astype(str)
        .str.replace(".", "", regex=False)
        .str.replace(",", ".", regex=False)
    )

    mines[col] = pd.to_numeric(
        mines[col],
        errors="coerce"
    )
```

```python
# Clean port numeric columns
port_num_cols = [
    "Latitude",
    "Longitude",
    "VesselInPort",
    "ExpectedArrival",
    "RecentPortCalls"
]

for col in port_num_cols:
    ports[col] = pd.to_numeric(
        ports[col],
        errors="coerce"
    )
```

```python
# Create month dictionary
bulan_map = {
    "Januari": "January",
    "Februari": "February",
    "Maret": "March",
    "April": "April",
    "Mei": "May",
    "Juni": "June",
    "Juli": "July",
    "Agustus": "August",
    "September": "September",
    "Oktober": "October",
    "November": "November",
    "Desember": "December"
}
```

```python
# Replace Indonesian month names
date_cols = [
    "Tanggal Berakhir SK",
    "Tanggal Berlaku SK"
]

for col in date_cols:
    mines[col] = mines[col].astype(str)

    for indo, eng in bulan_map.items():
        mines[col] = mines[col].str.replace(
            indo,
            eng,
            regex=False
        )

    mines[col] = pd.to_datetime(
        mines[col],
        errors="coerce"
    )
```

```python
# Categorical Type
cat_cols = [
    "Kecamatan",
    "Kabupaten",
    "Komoditi",
    "Kode Unsur",
    "Provinsi",
    "Jenis Izin",
    "CAPEX / OPEX",
    "Indikator Fatalitas",
    "Pembeli / Offtaker",
    "Jenis Kontrak",
    "Status Penyelidikan Mineral Logam"
]

for col in cat_cols:
    mines[col] = mines[col].astype("category")
```

```python
# Convert to datetime
for col in date_cols:
    mines[col] = pd.to_datetime(
        mines[col],
        errors="coerce"
    )
```

```python
mines[[
    "Tanggal Berakhir SK",
    "Tanggal Berlaku SK"
]].isnull().sum()
```

```python
mines[mines["Luas Wilayah"].isnull()]
```

```python
# Handle missing values
distance_fill = [
    "Mine_port(meter)",
    "Mine_road(meter)",
    "Mine_Smelter(meter)"
]

for col in distance_fill:
    mines[col] = mines[col].fillna(
        mines[col].median()
    )
```

FEATURE ENGINEERING

```python
# Add mining permit age
mines["License_Duration_Years"] = (
    mines["Tanggal Berakhir SK"] -
    mines["Tanggal Berlaku SK"]
).dt.days / 365
```

```python
# Add active permission status
today = pd.Timestamp.today()

mines["License_Status"] = np.where(
    mines["Tanggal Berakhir SK"] > today,
    "Active",
    "Expired"
)
```
