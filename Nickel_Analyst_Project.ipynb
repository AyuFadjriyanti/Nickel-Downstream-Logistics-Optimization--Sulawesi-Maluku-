DATA CLEANING

```python
# Clean the column
mines.columns = mines.columns.str.strip()
ports.columns = ports.columns.str.strip()
```

```python
# Cleaning mine number column
big_number_cols = [
    "Bijih Terbukti (Ton)",
    "Logam Terbukti (Ton)"
]

for col in big_number_cols:
    mines[col] = (
        mines[col]
        .astype(str)
        .str.replace(".", "", regex=False)
        .str.replace(",", ".", regex=False)
    )

    mines[col] = pd.to_numeric(
        mines[col],
        errors="coerce"
    )
```

```python
# Cleaning the port number column
coord_cols = [
    "longitude",
    "latitude"
]

for col in coord_cols:
    mines[col] = pd.to_numeric(
        mines[col],
        errors="coerce"
    )
```

```python
port_num_cols = [
    "Latitude",
    "Longitude",
    "VesselInPort",
    "ExpectedArrival",
    "RecentPortCalls"
]

for col in port_num_cols:
    ports[col] = pd.to_numeric(
        ports[col],
        errors="coerce"
    )
```

```python
# Distance columns
distance_cols = [
    "Mine_port(meter)",
    "Mine_road(meter)",
    "Mine_Smelter(meter)",
    "Luas Wilayah"
]

for col in distance_cols:
    mines[col] = pd.to_numeric(
        mines[col],
        errors="coerce"
    )
```

```python
# Cleaning date
date_cols = [
    "Tanggal Berakhir SK",
    "Tanggal Berlaku SK"
]

for col in date_cols:
    mines[col] = pd.to_datetime(
        mines[col],
        errors="coerce"
    )
```
